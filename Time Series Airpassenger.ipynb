```python
# We will Check if our series is Stationary or not
import matplotlib.pyplot as plt
plt.plot(a)
```

```python
#The series which we got is not Stationary 
a.diff()
```

```python
plt.plot(a.diff().diff().diff().diff().diff())
```

```python
# Even after multiple level of differencing series does not become stationary
# We are going to take log of the series
a_log=np.log(a)
```

```python
a_log
```

```python
plt.plot(a_log)
```

```python
plt.plot(a_log.diff().diff())
```

```python
# After taking log and applying the differencing we get the series Stationary
# Variance Stable krne ke liye log lete hai
# for building a model we need 3 things are p,q,d ,d=2
```

```python
# To find p and q we need to use ACF and PACF
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_acf(a)
```

```python
plot_pacf(a)
```

```python
# When your ACF declines slowly or gradually and pacf declines sharply then we called as AR Process
# After a 2 lags we are getting corelation=0 so this is AR(2) process
# AR mapped to P
# The values we got (p=2,d=2,q=0)
```

```python
# This will give Error as ARIMA is old and we will replace it by Sarima
from statsmodels.tsa.arima_model import ARIMA
model=ARIMA(a_log.Passengers,order=(2,1,0))
```

```python

# How to use Sarima and Sarimax
# We will use auto_arima and whatever vlaue we will get from auto arima we are going to pass it in our model
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
import pmdarima
from pmdarima import auto_arima
```

```python
result=seasonal_decompose(a_log)
result.plot();
```

```python
# This plot is just to get the idea about data 
auto_arima(a_log,seasonal=True,m=12)
# m is time period
```

```python
# Whatever values will get from auto_arima we will pass them in Sarimax
sarimax=SARIMAX(a_log,order=(2,0,0),seasonal_order=(0,1,1,12))
```

```python
a_=sarimax.fit()
```

```python
pred=a_.predict(start=144,end=155)
```

```python
pred
```

```python
pred1=np.exp(pred)
pred1
```

```python
plt.plot(a_log)
plt.plot(pred)
```
